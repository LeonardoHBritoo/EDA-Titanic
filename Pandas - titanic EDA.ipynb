### Incluindo coluna de idade nos cÃ¡lculos

```python
# MÃ©dia de idade das pessoas que sobreviveram por classe e sexo
df.groupby(['sex','pclass','survived'])[['age']].agg(['mean','count'])
```

```python
df['age'].isnull().sum() # Precisamos limpar a base antes
```

```python
#Quantos passageiros hÃ¡ em cada faixa etÃ¡ria? de 15 em 15 anos
def calc_faixa_etaria(idade):
    if np.isnan(idade): # Verifica se a idade Ã© NaN (Valor vazio ou invÃ¡lido)
        return 'Desconhecido'
    elif idade <= 15:
        return '[0 - 15]'
    elif idade <= 30:
        return '[15 - 30]'
    elif idade <= 45:
        return '[30 - 45]'  
    else:
        return '[45+]'

df['age'].apply(calc_faixa_etaria)
```

```python
# Atribuindo a uma nova coluna
df['faixa_etaria'] = df['age'].apply(calc_faixa_etaria)
df
```

```python
df['faixa_etaria'].value_counts()
```

### Vamos explorar a coluna de nomes para ver se os personagens do filme aparecem aqui


```python
df['name'] # Vendo os padrÃµes e estruturas da coluna
# Sobrenome, tÃ­tulo. primeiros nomes
```

```python
# Pegando sobrenomes
def pega_sobrenome(nome):
    return nome.split(',')[0]
df['sobrenome'] = df['name'].apply(pega_sobrenome)
```

```python
# Pegando tÃ­tulo
def pega_titulo(nome):
    return nome.split(',')[1].split('.')[0]
df['titulo'] = df['name'].apply(pega_titulo)

```

```python
# Pegando Primeiros nomes
def pegando_primeiros_nomes(nome):
    return nome.split(',')[1].split('.')[1]
df['Primeiros Nomes'] = df['name'].apply(pegando_primeiros_nomes)
```

```python
# TambÃ©m Ã© possÃ­vel usar funÃ§Ãµes lambda para os tratamentos acima, exemplo:
df['name'].apply(lambda nome : nome.split(',')[0]) # Pegando sobrenome com lambda
df['name'].apply(lambda nome : nome.split(',')[1].split('.')[0]) # pegando tÃ­tulo com lambda
df['name'].apply(lambda nome : nome.split(',')[1].split('.')[1]) # pegando primeiros nomes com lambda
```

```python
# Contando quantos sobrenomes repetidos temos
df['sobrenome'].value_counts().value_counts() # Basicamente conta as repetiÃ§Ãµes das repetiÃ§Ãµes, quantas pessoas tÃªm
# sobrenomes Ãºnicos, quantas tem sobrenomes que se repetem 2 vezes, 3 vezes.... 11 vezes
```

```python
df['titulo'].value_counts() # Decobrindo os tÃ­tulos dos passageiros e a distribuiÃ§Ã£o
```
